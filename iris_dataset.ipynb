#### Step 1: Import the required libraries


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Step 2: Load the Iris dataset

```python
iris = sns.load_dataset('iris')
iris.to_csv('iris.csv', index=False)

```

```python
iris = pd.read_csv('iris.csv')
```

#### Step 3: Inspect the dataset structure

```python
# Check the shape (rows, columns)
print("Shape of dataset:", iris.shape)
```

```python
# Display column names
print("Column names:", iris.columns)
```

```python
# Display first 5 rows
print(iris.head())
```

```python
# Display last 5 rows
print(iris.tail())
```

```python
iris.info()
```

#### Step 4: Basic Visualizations

##### Scatter plot to analyze relationships between variables. 

```python
# Create pair plot (scatter plots between all numeric variables)
sns.pairplot(iris, hue='species')
plt.suptitle('Scatter Plot Matrix of All Variables in Iris Dataset', y=1.02)
plt.show()
```

#####  Histogram to examine data distribution. 

```python
# Create histograms for all numeric columns
iris.hist(figsize=(10, 8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of All Numeric Variables in Iris Dataset', y=1.02)
plt.show()
```

#####  Box plot to detect outliers and spread of values. 

```python
# Automatically select numerical columns
numerical_cols = iris.select_dtypes(include=['int64', 'float64']).columns

# Create box plots for all numerical columns
plt.figure(figsize=(12,6))
iris[numerical_cols].boxplot()
plt.title("Box Plot of All Numerical Columns")
plt.show()
```

```python
# Box plot of sepal_length by species of categorical values
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Sepal Length by Species")
plt.show()
```
